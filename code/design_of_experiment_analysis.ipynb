### Design of Experiments - DoE Shiny Application
#### Modelling Simulation

This may require determining which variables are significant, assuming a model for the system.

#### Visualizing distribution of experimental data

Since I used Python to analyze the data, I referred to linear regression method [Ordinary Least Squares - OLS](https://www.statsmodels.org/stable/generated/statsmodels.regression.linear_model.OLS.html) to implement the model.
#### Creating linear regression model

```python
# Linear regression for each 'x' variable
for column in input_columns:
    X = df[[column]]
    y = df[output_column]

    # Adding a constant term to the input for statsmodels
    X = sm.add_constant(X)

    # Fitting the linear regression model
    model = sm.OLS(y, X).fit()

    # Display the summary statistics
    print(f"Summary for {column}:")
    print(model.summary())
    
    # Scatter plot with regression line
    plt.figure(figsize=(7, 5))
    sns.scatterplot(x=df[column], y=df[output_column])
    sns.lineplot(x=df[column], y=model.predict(X), color='red', label='Linear Regression')
    plt.title(f"{column} vs. {output_column}")
    plt.xlabel(column)
    plt.ylabel(output_column)
    plt.xlim(0, 1.1)
    plt.legend()
    plt.show()

```
