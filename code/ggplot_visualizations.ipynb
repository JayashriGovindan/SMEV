```python
import pandas as pd 
from plotnine import *
df = pd.read_csv("set1_flat.csv") 
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative' ) + geom_point() 
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative') + geom_point( alpha = 0.4) + theme_bw()
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative' ) + geom_jitter(alpha = 0.5, width= 0.2)+ theme_bw()
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative' ) + geom_jitter(alpha = 0.5, width= 0.2)+ theme_bw() + geom_line()
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative' ) + geom_jitter(alpha = 0.5, width= 0.2)+ theme_bw() + geom_boxplot(width=0.3)
```

```python
ggplot(df) + aes(x="Alternative", y="Time", color = 'Alternative' ) + theme_bw() + geom_errorbar ()
```

```python
ggplot(df) + aes(x="Time") + facet_wrap("Alternative")+theme_bw() + geom_histogram()
```
