### A) Analysis

1. Initially, an analysis was conducted on the initial name "Bertrand" in order to have a better understanding of the data and to become familiar with the exploration of data. Later, the entire dataset was subjected to some analysis.

4. The Y-axis, or year series, is not displayed precisely due to the large amount of data. Nonetheless, a distinct picture of the occurrences of the name rising and falling is evident.
5. Used a geom point plot to analyze the year widely, were the name occurrence reaches its maximum value.

Compared to the other two names, the name Fabrice originated later but had the highest frequency.All of these names, nevertheless, appear to have become less common recently.
9. An analysis was conducted on the most commonly given name by year. _PRENOMS_RARES, ENZO, Gabriel, Leo, Lucas, Mael, Nathan, Noe, Nolan, and Raphael are the ten names for whom the maximum count of occurrences was 100. ### Loading libraries

```r
library(tidyverse)
library(ggplot2)
FirstNames <- read_delim("dpt2021.csv",delim=";")
```

### Analysing data for the choosen first name "BERTRAND"

```r
#Analysing for the first name BERTRAND
firstname_bertrand = FirstNames  %>%filter(preusuel=="BERTRAND" & annais != 'XXXX' ) #Removing outlier where annais = XXXX
print(head(firstname_bertrand))
```

```r
# Checking the name occurence with respective to sexe
gender_specific = firstname_bertrand  %>%count(sexe)
print("Count of name for each gender:")
print(gender_specific)
```

```r
# Analysing year wise count for the name
bertrand_analysis = firstname_bertrand %>%count(annais)
bertrand_analysis
```

### Plotting data for frequency of name "BERTRAND" along time (year)

```r
ggplot(bertrand_analysis, aes(x = annais, y = n,group = 1)) +  geom_line() + ggtitle("Frequency of name 'BERTRAND' along time")
```

### Visualizing frequency of name Bertand over the years

```r
ggplot(bertrand_analysis, aes(x = annais, y = n, color=factor(annais)))+ geom_point() + ggtitle("Occurence of name 'Bertrand' over years")
```

###  Finding the maximum number of name occurences and the year in which it occurs

```r
print("Maximum number of name occurences with respect to year: ")
print(max(bertrand_analysis$n))
print("Year with maximum number of occurences: ",)
bertrand_analysis_max = FirstNames  %>%filter(preusuel=="BERTRAND", annais>=1970 & annais<=1980) %>%count(annais)
ggplot(data = bertrand_analysis_max, aes(x=annais, y=n,fill = annais )) + geom_bar(stat="identity") + ggtitle("Data of Bertrand from 1970 - 1980")
bertrand_analysis_max = bertrand_analysis_max  %>%filter(n == max(bertrand_analysis$n)) 
print(bertrand_analysis_max)
```

### Checking geographical impact on name "Bertrand"

```r
#Analysing department wise count for the name
bertrand_analysis_dept = FirstNames  %>%filter(preusuel=="BERTRAND" ) %>%count(dpt) 
print(bertrand_analysis_dept)
```

```r
ggplot(bertrand_analysis_dept, aes(x = dpt, y = n, group=1))+ geom_line() + geom_point() + ggtitle("Department wise name 'Bertrand' occurences")
print("Maximum number of name occurences with respect to department: ")
print(max(bertrand_analysis_dept$n))
print("Department with highest number of name occurence")
bertrand_analysis_dept= bertrand_analysis_dept  %>%filter(n == max(bertrand_analysis_dept$n)) 
print(bertrand_analysis_dept)
```

### Comparing different first name frequencies along time

```r
arthur_analysis = FirstNames  %>%filter(preusuel=="ARTHUR") %>%count(annais)
fabrice_analysis = FirstNames  %>%filter(preusuel=="FABRICE") %>%count(annais)
kleber_analysis = FirstNames  %>%filter(preusuel=="KLEBER") %>%count(annais)
```

### Most given firstname by year along with its frequency

```r
ggplot() +
geom_line(data=arthur_analysis,aes(x = annais, y = n, group = 1), color='blue',) + 
geom_line(data=fabrice_analysis,aes(x = annais, y = n, group = 1), color='red') + 
geom_line(data=kleber_analysis,aes(x = annais, y = n, group = 1), color='green') +
ggtitle("Comparing different first name frequencies along time")
```

```r
most_given_name = FirstNames  %>%filter(sexe=='1') %>%group_by(annais) %>% count(preusuel)
most_given_name = most_given_name  %>%filter(n == max(most_given_name$n)) 
print("Most given firstname by year:")
print(most_given_name)
ggplot(most_given_name, aes(x = preusuel, y = annais ,group = 1, color=factor(preusuel))) +  geom_line() + geom_point() + ggtitle("Frequency of most given firstname by year")
```
